```python
from iertools.read import *
from dateutil.parser import parse
```

```python
f = 'data/Casablanca.epw' # Archivo con datos horarios
casablanca = read_epw(f,year=2000)
casablanca.plot(subplots=True,figsize=(12,20));
```

```python
f = 'data/Tmx_marzoCST.epw' # Archivo de 1 semana de marzo con datos cada 10 minutos
tmx = read_epw(f,alias=True)
tmx.plot(subplots=True,figsize=(12,20));
```

```python
f = 'idf/cubo.sql'
idf = read_sql(f).data
```

```python
idf
```

```python
fig, ax = plt.subplots(figsize=(12,3))
f1 = parse('2018-03-9')
f2 = f1 + pd.Timedelta('1D')

ax.plot(idf['TECHO:Surface Outside Face Incident Solar Radiation Rate per Area (W/m2)'],'ko')
ax.plot(tmx.Ig,'r.')

ax.set_xlim(f1,f2)
```

```python
fig, ax = plt.subplots(figsize=(12,3))
f1 = parse('2018-03-9')
f2 = f1 + pd.Timedelta('1D')

ax.plot(idf['Environment:Site Wind Speed (m/s)'],'ko',label='Simulacion')
ax.plot(tmx.Ws,'r.',label='epw')

ax.set_xlim(f1,f2)
ax.legend()
```
